Import Pandas


```python
import pandas as pd
```

Read the `pokemon.csv` file and set the index to `['Type 1', 'Name']`


```python
pokemons = pd.read_csv("Pokemon.csv", index_col=["Type 1", "Name"])
pokemons.head()
```

Sort the pokemons by `Type 1`


```python
pokemons.sort_values("Type 1", inplace=True)
```

Select all the `Fire` pokemons that start with the letter `C`


```python
fire_pokemon = pokemons.loc["Fire"]
start_with_c = fire_pokemon.index.str.startswith("C")
fire_pokemon[start_with_c]
```

Using `agg()` get the `max`, `min` and the `mean` of all the pokemons for the columns `['Attack', 'Defense']` and round them to 1 decimal place.


```python
pokemons.agg(
    {"Attack": ["max", "min", "mean"], "Defense": ["max", "min", "mean"]}
).round(1)
```

Create a function that checks the Attack of a pokemon. If the `Attack` is over 100, it should now display `'STRONG'`, if it is less than that, it should display `'WEAK'`. Update the DataFrame to display this new Attack column`


```python
def strength(attack):
    if attack > 100:
        return "STRONG"
    else:
        return "WEAK"


pokemons["Strength"] = pokemons["Attack"].apply(strength)
```

Create a new column called `Reversed` which holds each pokemon name reversed


```python
# pokemons["Reversed"] = pokemons.index.get_level_values(1).str[::-1]

pokemons["Reversed"] = pokemons.index.get_level_values(1)


def reverse(string: str) -> str:
    return string[::-1]


pokemons["Reversed"].apply(reverse)
```

Reset the index so that only `Name` shows as the index


```python
pokemons.reset_index(0, drop=False, inplace=True)
```
